Next stage is to get familiar with salary_range values:

```python
canada_jobs['salary_range'].unique()
```

It is necessary to mark certain rows with no salary provided, so we are able to slice the data, based on this predicate

```python
canada_jobs['salary_range'] = canada_jobs['salary_range'].fillna('-1')
```

As we can see there are two noticeable patterns the US table and one in Canada's. Thus, I suggest to retrieve the meaningful values from the jobs in the country, whose symbol is clover, at first.

At first, we filter out the rows, where salary is not provided, then replace all capital letters "K" after the number with three zeroes to indicate thousands, remove dollar sign before actual numbers and truncate the text part at all

```python
clean_salary_canada = canada_jobs[canada_jobs['salary_range'] != '-1']
```

```python
#Ensure there is no missing values
clean_salary_canada['salary_range'].isna().sum()
```

```python
clean_salary_canada['salary_range'] = clean_salary_canada['salary_range'].apply(lambda x:x.replace('K', '000').replace('$', '').split('(')[0])
```

Now we get a dummy variable to indicate the offerings, where salaries are shown per hour . It will help us in future calculations

```python
clean_salary_canada['per_hour_variable'] = clean_salary_canada['salary_range'].apply(lambda x: 1 if 'per hour' in str(x).lower() else 0)
```

Now we can get rid of that string as well:

```python
clean_salary_canada['salary_range'] = clean_salary_canada['salary_range'].apply(lambda x:x.split(' P')[0])
```

Finally, we are able to split the range into minimal and maximum wage per year(we will convert hour rates later):

```python
#clean_salary_canada['salary_range'] = clean_salary_canada['salary_range'].replace('nan', 'none - none')
#canada_jobs['salary_range'] = canada_jobs['salary_range'].replace(-1, 'none - none')
```

```python
clean_salary_canada['min_salary'] = clean_salary_canada['salary_range'].apply(lambda x:x.split('-')[0])
```

```python
#Looking at unique values of salary range I noticed some employers provided only one certain wage
#So in the process of indentifying maximal salary the code was wrapped in if-condition
for index, row in clean_salary_canada.iterrows(): 
    if '-' in row['salary_range']:
        clean_salary_canada.at[index, 'max_salary'] = row['salary_range'].split('-')[1]
    else:
        clean_salary_canada.at[index, 'max_salary'] = row['min_salary']
```

```python
clean_salary_canada.head(50)
```

At this point we shall declare newly created columns as of float data type:

```python
clean_salary_canada['min_salary'] = clean_salary_canada['min_salary'].astype('float')
clean_salary_canada['max_salary'] = clean_salary_canada['max_salary'].astype('float')
```

Now we are going to turn all values into salary per year. That is where 'per hour' variable will come in handy

```python
#According to the source(https://snappyrates.ca/working-days-in-canada/), 
#there are 173 working hours per month on average, consequently 2080 hours
#this number will be taken in the computation
for index, row in clean_salary_canada.iterrows(): 
    if row['per_hour_variable'] == 1:
        clean_salary_canada.at[index, 'min_salary'] = row['min_salary']*2080
        clean_salary_canada.at[index, 'max_salary'] = row['max_salary']*2080
```

Additionally, while parsing the unique values of saly range field, I saw  that several records contained estimated salary per month next couple of lines of code are going to fix the issue


```python
#We just multiply the salary on the amount of months in a year
for index, row in clean_salary_canada.iterrows(): 
    if (row['min_salary'] >= 1000) and (row['min_salary'] <= 15000):
        clean_salary_canada.at[index, 'min_salary'] = row['min_salary']*12
        clean_salary_canada.at[index, 'max_salary'] = row['max_salary']*12

```

Let's check the result of the manipulation:

Everything looks just fine as it is supposed to do, except maximum value it seems strange. We should examine the record with that salary and apply for that job immediately:

```python
#Just in case, we drop duplicates one more time to prevent any redundancy
clean_salary_canada = clean_salary_canada.drop_duplicates(subset = ['job_title', 'company_name'])
```

```python
#clean_salary_canada.loc[clean_salary_canada['max_salary'] == np.max(clean_salary_canada['max_salary'])]
clean_salary_canada.sort_values('max_salary', ascending = False).head(10)
```

It still looks weird, but we see there is a phrase 'part-time' in the title of the job. Thus, supposedly the calculated maximum and minimal wage are hard to reach, but we chose the amount of working hours to be met by full-time positions(2080 hours per year). For the part-time roles this number is equivocal. The authorized source Indeed(https://ca.indeed.com/career-advice/finding-a-job/how-many-hours-is-part-time) states that in Canada any amount of hours worked less than 30 is considered as part-time. Glassdoor(https://www.glassdoor.com/blog/guide/how-many-hours-is-part-time/), however, defines a quite wide but more certain range - 20 to 29 hours. We are going to choose the middle number, which, in this occassion, is 25 hours. Some of not hard maths and we get 1300 hours per year. This variable will be used in part-time salary calculations.

```python
for index, row in clean_salary_canada.iterrows():
    if  ('part time' in row['job_title'].lower()) or ('part-time' in row['job_title'].lower()):
        clean_salary_canada.at[index, 'min_salary'] = (row['min_salary']/2080)*1300 #we divide by 2080 to get back to hour rate
        clean_salary_canada.at[index, 'max_salary'] = (row['max_salary']/2080)*1300
```

And to make salary attribute comparable with the USA jobs data, we convert the salaries into US dollars:

```python
#currency rate on 01.02.2023: 1 CAD = 0,74356 USD
clean_salary_canada['min_salary'] = clean_salary_canada['min_salary']*0.74
clean_salary_canada['max_salary'] = clean_salary_canada['max_salary']*0.74
```
