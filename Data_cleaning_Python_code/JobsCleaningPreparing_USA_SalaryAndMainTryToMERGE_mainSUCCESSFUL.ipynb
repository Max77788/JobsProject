Let's begin defining the site the job offer was posted on, relying on the fields we have in access

```python
usa_jobs['site'] = usa_jobs['joblink_URL'].apply(lambda x: 'linkedin' if 'linkedin' in str(x).lower() else ('indeed' if 'indeed' in str(x).lower() else 'glassdoor')) 
```

I suggest to proceed with working out the 'location' field:

```python
#complete 'remote' feature engineering
usa_jobs['remote_locat'] = usa_jobs['location'].apply(lambda x: 'yes' if 'remote' in x.lower() else 'no')
usa_jobs['remote_job_tit'] = usa_jobs['job_title'].apply(lambda x: 'yes' if 'remote' in x.lower() else 'no')
usa_jobs['remote'] = usa_jobs.apply(lambda x: 'yes' if x['remote_job_tit'] == 'yes' else ('yes' if x['remote_locat'] == 'yes' else 'no'), axis=1)
usa_jobs.drop(columns = ['remote_locat', 'remote_job_tit'], axis=1,inplace=True)
usa_jobs[25:70]
```

```python
#pull out the city and state
usa_jobs['city'] = usa_jobs['location'].apply(lambda x:x.split('\n')[0].split(',')[0])
usa_jobs['state'] = usa_jobs['location'].apply(lambda x: x.split('\n')[0].split(',')[1][:3] if ',' in x else 'not indicated')
```

Still there are some misperfections in our own derived fields, such as 'remote' and 'United States' in 'city' column, however, these values are quite informative and can be handled easily during the exploratory analysis and visualization phases. Therefore we are done we these attributes

Next stage is to get familiar with salary_range values:

```python
usa_jobs['salary_range'].unique()
```

It is necessary to mark certain rows with no salary provided, so we are able to slice the data, based on this predicate

```python
usa_jobs['salary_range'] = usa_jobs['salary_range'].fillna('-1')
```

As we can see there are two noticeable patterns the US table and one in Canada's. Thus, I suggest to retrieve the meaningful values from the jobs in the country, whose symbol is clover, at first.

At first, we filter out the rows, where salary is not provided, then replace all capital letters "K" after the number with three zeroes to indicate thousands, remove dollar sign before actual numbers and truncate the text part at all

```python
clean_salary_usa = usa_jobs[usa_jobs['salary_range'] != '-1']
```

```python
#Ensure there is no missing values
clean_salary_usa['salary_range'].isna().sum()
```

```python
clean_salary_usa['salary_range'] = clean_salary_usa['salary_range'].apply(lambda x:x.replace('K', '000').replace('US$', '').replace('$', '').replace(',','').replace('£','').split('(')[0].split(' P')[0])
```

```python
clean_salary_usa['salary_range'].unique()
```

```python
clean_salary_usa['min_salary'] = clean_salary_usa['salary_range'].apply(lambda x:x.split('-')[0].strip())
clean_salary_usa['min_salary'] = clean_salary_usa['min_salary'].astype('float')
```

```python
#clean_salary_usa['max_salary'] = clean_salary_usa.apply(lambda x:x['salary_range'].split('-')[1].strip() if '-' in x else , axis=1)
```

```python
clean_salary_usa['max_salary_test'] = clean_salary_usa['salary_range'].apply(lambda x:x[-7:])
clean_salary_usa['max_salary_test2'] = clean_salary_usa['max_salary_test'].apply(lambda x:x.replace('-','').strip())
clean_salary_usa['max_salary'] = clean_salary_usa['max_salary_test2'].apply(lambda x: x[-3:].strip() if ' ' in x else x.strip())
clean_salary_usa['max_salary'] = clean_salary_usa['max_salary'].astype('float')
clean_salary_usa.drop(columns = ['max_salary_test', 'max_salary_test2'], axis = 1, inplace=True)
```

```python
#converting from hour wage to year one, taking 2080 an amount of working hours in 2023
clean_salary_usa['max_salary'] = clean_salary_usa['max_salary'].apply(lambda x: x*2080 if x<150 else x)
clean_salary_usa['min_salary'] = clean_salary_usa['min_salary'].apply(lambda x: x*2080 if x<150 else x)
```

```python
clean_salary_usa[100:150]
```

```python
clean_salary_usa = clean_salary_usa.drop_duplicates(subset = ['job_title', 'company_name'])
```

```python
clean_salary_usa.describe()
```

In general, all fields contain adequate values. The only suspicious part is minimal value of max and min salary. Let's take a closer look at it:

```python
clean_salary_usa.sort_values('max_salary').head(10)
```

There are three rows, missing the meaningful value for salarie column. In order to fix this we are going to replace it with NaN values, so there will be no misinterpretation during EDA and visualization stages

```python
clean_salary_usa['min_salary'] = clean_salary_usa['min_salary'].apply(lambda x: x if x != 0 else np.nan)
clean_salary_usa['max_salary'] = clean_salary_usa['max_salary'].apply(lambda x: x if x != 0 else np.nan)
```

```python
clean_salary_usa.describe()
```

```python
clean_salary_usa.sort_values('min_salary').head(10)
```

As it turns out there is one posting, where the salary is indicated per month. Let's change it quickly:

```python
clean_salary_usa['min_salary'] = clean_salary_usa['min_salary'].apply(lambda x: x*12 if (x > 5000 and x < 10000) else x)
clean_salary_usa['max_salary'] = clean_salary_usa['max_salary'].apply(lambda x: x*12 if (x > 5000 and x < 10000) else x)
```

```python
clean_salary_usa.describe()
```

```python
cleaned_usa_df = usa_jobs.merge(clean_salary_usa, on=['job_title','company_name'], how='outer', suffixes=['', '_clean'])
```

```python
columns_to_drop = ['joblink_URL_clean', 'location_clean',
       'salary_range_clean', 'rating_clean', 'emp_type_clean',
       'job_level_clean', 'job_responsib_clean', 'industry_clean',
       'site_clean', 'remote_clean', 'city_clean', 'state_clean']
cleaned_usa_df = cleaned_usa_df.drop(columns_to_drop, axis=1)
```
